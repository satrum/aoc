```python
datalist = [string for string in data.split('\n') if len(string)>0]
```

```python
len(datalist[0]), len(datalist) #start 0,0 right 3 , down 1 
```

```python
width = len(datalist[0])
height = len(datalist)

def get_count(x_delta, y_delta):
    x = 0
    y = 0
    count = 0
    while y<height:
    #for i in range(height):
        if datalist[y][x] == '#':
            count += 1
        #print(i, x,y,datalist[y][x], count)
        x += x_delta
        y += y_delta
        if x>=width:
            x -= width
    return count

options = [
    (1,1),
    (3,1),
    (5,1),
    (7,1),
    (1,2)
]
m = 1
for option in options:
    count = get_count(*option)
    print(count)
    m *= count
print(m)

#print(x, y)
```

```python
file = 'adv4.txt'
with open(file, 'r') as f:
    data = f.read()
```

```python
datalist = [string for string in data.split('\n')]
```

```python
passport = {}
passports = []
for item in datalist:
    if item == '':
        passports.append(passport)
        passport = {}
    else:
        for string in item.split(' '):
            passport[string.split(':')[0]] = string.split(':')[1]
```

```python
passports
```

```python
valid = [
    'byr','iyr','eyr','hgt','hcl','ecl','pid',#'cid'
]
valid_count = 0
for passport in passports:
    valid_keys = 0
    for key in valid:
        if key in passport.keys():
            valid_keys += 1
    if valid_keys == 7:
        hgt = int(passport['hgt'][:-2])
        hgt_s = passport['hgt'][-2:]
        try:
            pid_flag = int(passport['pid'])>=0
        except Exception:
            pid_flag = False
        hcl = [i for i in passport['hcl'] if i in '0123456789abcdef']
        
        if int(passport['byr'])>2002 or int(passport['byr'])<1920:
            print('byr: ', passport['byr'])
        elif int(passport['iyr'])>2020 or int(passport['iyr'])<2010:
            print('iyr: ', passport['iyr'])
        elif int(passport['eyr'])>2030 or int(passport['eyr'])<2020:
            print('eyr: ', passport['eyr'])
        elif hgt_s == 'cm' and (hgt>193 or hgt<150):
            print('hgt: ', passport['hgt'])
        elif hgt_s == 'in' and (hgt>76 or hgt<59):
            print('hgt: ', passport['hgt'], hgt)
        elif (hgt_s != 'in') and (hgt_s != 'cm'):
            print('hgt: ', passport['hgt'], hgt_s)
        #hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
        elif (len(passport['hcl'])!=7) or (passport['hcl'][0]!='#'):
            print('hcl: ', passport['hcl'])
        elif len(hcl) !=6:
            print('hcl: ', hcl)
        elif passport['ecl'] not in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']:
            print('ecl: ', passport['ecl'])
        elif len(passport['pid'])!=9:
            print('pid: ', passport['pid'], len(passport['pid']))
        elif not pid_flag:
            print('pid: ', passport['pid'], pid_flag)
        else:
            valid_count += 1
            #print(hcl)
    else:
        pass
        #print(passport)
```
